```python
# The following block contains code to merge the two data sets and clean data
```

```python
import pandas as pd

from_redfin = pd.read_csv('redfin_sold_3mo.csv')
scrape_redfin = pd.read_csv('redfin_image_url.csv')
```

```python
from_redfin.head()
from_redfin.shape
```

```python
scrape_redfin.head()
scrape_redfin.shape
# size is different because scraper failed to pick up 3 records
```

```python
merged_inner = pd.merge(left=from_redfin, right=scrape_redfin, left_on="ADDRESS", right_on="address")
merged_inner.shape
# inner join omits the 3 records that were not scraped
```

```python
merged_inner #need to clean data
```

```python
merged_inner['SOLD DATE'].isnull()
```

```python
asdf=merged_inner[merged_inner['SOLD DATE'].isnull()==False] #removed records without days on market field
```

```python
asdf1=asdf[asdf['thumbnail-src'].isnull()==False] #removed records without valid thumbnail
asdf1
```

```python
townhouse = asdf1[asdf1['PROPERTY TYPE']=="Townhouse"]
condo = asdf1[asdf1['PROPERTY TYPE']=="Condo/Co-op"]
sfh = asdf1[asdf1['PROPERTY TYPE']=="Single Family Residential"]
```

```python
townhouse.to_csv('townhouse.csv', index=False) 
condo.to_csv('condo.csv', index=False) 
sfh.to_csv('sfh.csv', index=False) 
```
